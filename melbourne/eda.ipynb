```python
# Converting data in numerical and categorical
numerical_columns = ['Price', 'Distance','Postcode', 'Bedroom2', 'Bathroom', 'Car', 'Landsize',
                     'BuildingArea','YearBuilt','Lattitude','Longtitude','Propertycount']
for column in numerical_columns:
    df[column] = df[column].astype(float)

categorical_columns = ['CouncilArea','Regionname',]
for column in categorical_columns:
    df[column] = df[column].astype('object')
```

```python
# Handling the missing values of the numerical and categorical data

for column in numerical_columns:
    df[column].fillna(df[column].median(), inplace=True)# 

for column in categorical_columns:
    df[column].fillna(df[column].mode()[0], inplace=True)
```

## outlier handling

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.boxplot(x=df["BuildingArea"],orient="h")
plt.xlabel('name')
plt.ylabel('count')
plt.show()

```

```python
# short method to handle outlier
for cols in df.columns[::]:
    if df[cols].dtype == object:
        continue
    else:
        Q1 = df[cols].quantile(q=0.25)
        Q3 = df[cols].quantile(q=0.75)
        iqr = Q3-Q1
    
        low = Q1-1.5*(Q3-Q1)
        high = Q3+1.5*(Q3-Q1)
        df.loc[(df[cols]<low) | (df[cols]>high),cols]=df[cols].median()
```
